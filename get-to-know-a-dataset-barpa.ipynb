#### BARPA-R Convective Parameters (BARPA/output-CMIP6/DD/AUST-15)
BARPA-R is a moderate-resolution (17 km) regional climate projections over Australia and the surrounding regions including parts of southeast Asia and New Zealand.
The dataset includes 3-hourly data for a collection of convective parameters - diagnostics used to characterise environments favourable for thunderstorms and their associated hazards (heavy rainfall, large hail, damaging wind gusts, and tornadoes) - over Australia. These include: parcel-based indices such as convective available potential energy (CAPE) and convective inhibition (CIN) for surface-based, mixed-layer (ML), and most-unstable (MU) parcels; thermodynamic parameters such as bulk wind difference (BWD); and composite parameters such as derecho composite (DCP), significant tornado parameter (STP). Daily and monthly mean and maximum values are also available for surface-based CAPE and CIN.

#### BARPA-R (BARPA/output-CMIP6/DD/AUS-15)
BARPA-R is a moderate-resolution (17 km) regional climate projections over Australia and the surrounding regions including parts of southeast Asia and New Zealand. The data set includes hourly, 3-hourly, daily and monthly data for temperature, moisture, wind and flux variables at sub-surface, surface, and pressure levels, and heights above surface. The vertical levels include 1000, 950, 925, 850, 700, 600, 500, 400, 300, 250, 200, 150, 100, 70, 50, 30, 20, 10 hPa pressure levels (18) and 10, 50, 100, 150, 200, 250 and 1500 metre above surface.

#### BARPA-C - Convective-Scale (BARPA/output-CMIP6/DD/AUST-04)
BARPA-C is a kilometre-scale (4.4 km) regional climate projection, over Australia.
The data set includes 20-minute, hourly, 3-hourly, daily and monthly data for temperature, moisture, wind and flux variables at sub-surface, surface, and pressure levels , and heights above surface. There are additional variables unique to convection-permitting models, namely number of lightning flash, updraft helicity, column-maximum updraft, and derived radar reflectivity. The vertical levels include 1000, 975, 950, 925, 900, 850, 800, 750, 700, 600, 500, 400, 300, 250, 200, 150, 100, 70, 50, 30, 20, 10 hPa pressure levels (22) and 50, 100, 150, 200, 250 and 1500 metre above surface.

#### CORDEX-CMIP6-BARPA-R (BARPA/CORDEX-CMIP6/DD/AUS-20i)
BARPA-R is a moderate-resolution (17 km) regional climate projections over Australia and the surrounding regions including parts of southeast Asia and New Zealand - CORDEX-Australasia region.
The model is registered with source_id BARPA-R1-NN.
A subset of BARPA-R modelling variables are regridded for CMIP6-CORDEX standard grid AUS-20i. These variables belong to the CMIP6-CORDEX ATMOS CORE variables. They include hourly, daily and monthly data for surface and near-surface variables for pressure, temperature, moisture, wind, precipitation and radiation.

### Q: What data formats are present in your dataset? What kinds of data are stored using these formats? Can you give any advice for how you work with these data formats?

Our datasets come as a large number of netCDF4 files. Working with netCDF (.nc) files involves handling multidimensional scientific data. netCDF files are self-describing, holding variables, dimensions, and metadata (CF conventions), making them standard for climate and atmospheric data. Use libraries such as xarray and netCDF4. xarray  provides a high-level, pandas-like interface for multi-dimensional data, while the netCDF4 library offers a lower-level interface, offering full control over file structure, compression, and other advanced features. Data is typically accessed as NumPy arrays.

See https://docs.xarray.dev/en/stable/user-guide/io.html for loading NetCDF intto xarray.
