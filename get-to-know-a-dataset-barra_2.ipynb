#### BARRA-R2 Regional Version 2
BARRA-R2 is a moderate-resolution (12 km) regional atmospheric reanalysis over Australia and the surrounding regions including parts of southeast Asia and New Zealand. The data set includes hourly, 3-hourly, daily and monthly data for temperature, moisture, wind and flux variables at sub-surface, surface, and pressure levels, and heights above surface. The vertical levels include 1000, 950, 925, 850, 700, 600, 500, 400, 300, 250, 200, 150, 100, 70, 50, 30, 20, 10 hPa pressure levels (18) and 10, 50, 100, 150, 200, 250 and 1500 metres above surface.

Directory: *BARRA2/output/reanalysis/AUS-11/BOM/ERA5/historical/hres/BARRA-R2*

#### BARRA-R2 Convective Parameters
BARRA-R2 is a moderate-resolution (12 km) regional atmospheric reanalysis over Australia and the surrounding regions including parts of southeast Asia and New Zealand. The dataset includes hourly data for a collection of convective parameters - diagnostics used to characterise environments favourable for thunderstorms and their associated hazards (heavy rainfall, large hail, damaging wind gusts, and tornadoes) - over Australia. These include: parcel-based indices such as convective available potential energy (CAPE) and convective inhibition (CIN) for surface-based (SB), mixed-layer (ML), and most-unstable (MU) parcels; thermodynamic parameters such as precipitable water, saturation fraction, and lapse rates for various layers; and kinematic parameters such as bulk wind difference (BWD) and storm-relative helicity (SRH) for both fixed and "effective" layers. Daily and monthly mean and maximum values are also available for SBCAPE and SBCIN.

Directory: *BARRA2/output/reanalysis/AUST-11/BOM/ERA5/historical/hres/BARRA-R2*

#### BARRA-RE2 Regional Ensemble Version 2
BARRA-RE2 is a moderate-resolution (24 km) downscaled ensemble for BARRA-R2 generated with ERA5 perturbations, over Australia and the surrounding regions including parts of southeast Asia and New Zealand. The data set includes hourly, 3-hourly, daily and monthly data for temperature, moisture, wind and flux variables at sub-surface, surface, pressure levels, and heights above surface, for the 22 time-lagged member ensemble. The vertical levels include 1000, 950, 925, 850, 700, 600, 500, 400, 300, 250, 200, 150, 100, 70, 50, 30, 20, 10 hPa pressure levels and 10, 50, 100, 150, 200, 250 and 1500 metres above surface.

Directory: *BARRA2/output/reanalysis/AUS-22/BOM/ERA5/historical/eda/BARRA-RE2*

#### BARRA-RE2 Convective Parameters 
BARRA-RE2 is a moderate-resolution (24 km) downscaled ensemble for BARRA-R2 generated with ERA5 perturbation, over Australia and the surrounding regions including parts of southeast Asia and New Zealand. The dataset includes sub-daily data for a collection of convective parameters—diagnostics used to characterise environments favourable for thunderstorms and their associated hazards (heavy rainfall, large hail, damaging wind gusts, and tornadoes) - over Australia.

Directory: *BARRA2/output/reanalysis/AUST-22/BOM/ERA5/historical/eda/BARRA-RE2*

#### BARRA-C2 Convective-Scale Version 2
BARRA-C2 is a kilometre-scale (4.4 km) downscaled reanalysis, over Australia. The data set includes 20-minute, hourly, 3-hourly, daily and monthly data for temperature, moisture, wind and flux variables at sub-surface, surface, pressure levels, and heights above surface. There are additional variables unique to convection-permitting models, namely lightning flash rate, updraft helicity, and derived radar reflectivity. The vertical levels include 1000, 975, 950, 925, 900, 850, 800, 750, 700, 600, 500, 400, 300, 250, 200, 150, 100, 70, 50, 30, 20, 10 hPa pressure levels (22) and 10, 50, 100, 150, 200, 250 and 1500 metres above surface.
The data set also includes 27 convective parameters that characterise environments favourable for thunderstorms and their associated hazards (heavy rainfall, large hail, damaging wind gusts, and tornadoes). Caution is advised when using these data as convective parameters are more commonly computed from coarse-resolution reanalyses with parameterised convection. 

Directory: *BARRA2/output/reanalysis/AUST-04/BOM/ERA5/historical/eda/BARRA-C2*

### Q: What data formats are present in your dataset? What kinds of data are stored using these formats? Can you give any advice for how you work with these data formats?

The datasets are provided as extensive sets of netCDF4 files. Their structure is designed to align, where appropriate, with the CORDEX‑CMIP6 archival specification, enabling easier use of the reanalysis in regional climate research and related applications.
  
The datasets are organised as one data file per parameter (and per vertical pressure level, if relevant), per month, across temporal frequencies ranging from 20‑minute to monthly.

Working with netCDF (.nc) files involves handling multidimensional scientific data. netCDF files are self-describing, holding variables, dimensions, and metadata (CF conventions), making them standard for climate and atmospheric data. Use libraries such as xarray and netCDF4. xarray  provides a high-level, pandas-like interface for multi-dimensional data, while the netCDF4 library offers a lower-level interface, offering full control over file structure, compression, and other advanced features. Data is typically accessed as NumPy arrays.

See https://docs.xarray.dev/en/stable/user-guide/io.html for loading NetCDF into xarray.

Regional atmospheric reanalyses like BARRA2 offer a remarkably rich view of the atmosphere because they blend global observational networks with high resolution numerical modelling to reconstruct atmospheric conditions consistently through time. Unlike coarse global products, regional reanalyses resolve fine‑scale features from boundary‑layer structure and local wind systems to mesoscale convection and upper‑tropospheric dynamics—capturing how processes interact vertically and horizontally across the domain. The multi‑decadal coverage of BARRA2, from 1979, provides a physically coherent picture of temperature, moisture, winds, and pressure throughout the atmospheric column, enabling detailed analysis of weather systems, climate variability, and extremes. This richness makes regional reanalysis data an indispensable foundation for research, impact assessment, and applications that depend on understanding the full three‑dimensional behaviour of the atmosphere and land surface.

#### EXAMPLE 1: Study of the New South Wales Heatwave of January 2017

In January and February, there were three distinct heatwaves in southeast Australia, with the highest temperatures recorded over 9–12 February 2017.

There were 3 heatwaves across January and early February 2017 saw unusually high daily maximum and minimum temperatures for at least three consecutive days over large parts of the country. The first heatwave began around 10 January and continued to 14 January.

For much of January, the subtropical ridge included an area of sustained high pressure over the Tasman Sea (Tasman High). While a Tasman High does not always lead to warmer conditions, its position and dynamical characteristics often favour elevated temperatures across southeast Australia. Between 10 and 14 January, a slow‑moving high over the Tasman Sea combined with a trough across central Australia to maintain a hot northerly airflow over the eastern states, reinforcing the prolonged heat ([Bureau of Meteorology, 2017, Special Climate Statement 61 - Exceptional heat in southeast Australia in early 2017](https://www.bom.gov.au/climate/current/statements/scs61.pdf)).

````python
# Load required Python modules 
import os, sys
import glob
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import cartopy.crs as ccrs
from matplotlib.colors import ListedColormap, BoundaryNorm

# You may also need to install netcdf4 for xarray
# https://docs.xarray.dev/en/stable/user-guide/io.html#backend-selection
# https://pypi.org/project/netCDF4/
````

````python
# Simple method used to load datasets

# TODO: Update to S3 URI
s3_uri = 'REPLACE_WITH_URI/BARRA2/output/reanalysis'

def open_barra_dataset(model_id, domain_id, freq, variable, yearmonth, s3 = s3_uri):
    """
    Load a BARRA2 reanalysis dataset for a specified model, domain, variable,
    frequency, and year–month period.

    This function constructs the expected file path for a BARRA2 NetCDF file, 
    then opens the dataset with xarray. It assumes the
    file exists and follows the standard BARRA2 directory and naming
    conventions.

    Parameters
    ----------
    model_id : str
        Identifier for the model configuration (e.g., "BARRA-C2", "BARRA-R2").
    domain_id : str
        Spatial domain identifier used in the BARRA2 directory structure.
    freq : str
        Temporal frequency of the dataset (e.g., "1hr", "day").
    variable : str
        Variable name (e.g., "tas", "pr") used in both the directory and filename.
    yearmonth : str
        Year–month string in YYYYMM format specifying the target file period.

    Returns
    -------
    xarray.Dataset
        The opened BARRA2 dataset corresponding to the requested parameters.

    Notes
    -----
    This function does not perform error checking. If the constructed file path
    does not exist or the file is unreadable, `xr.open_dataset` will raise an
    exception.
    """
    root_datapath = f'{domain_id}/BOM/ERA5/historical/hres/{model_id}/v1/{freq}'
    file_path = os.path.join(
        s3,
        root_datapath,
        variable,
        'latest',
        f"{variable}_{domain_id}_ERA5_historical_hres_BOM_{model_id}_v1_{freq}_{yearmonth}-{yearmonth}.nc")
    
    ds = xr.open_dataset(file_path)
    
    return ds
````

````python
# Define which sub-collection of BARRA2
freq = '1hr'
yearmonth = '201701'

# Open the BARRA2 datasets for 4 near-surface meteorological parameters using xarray
#   and put them in a da_list dict as xr.DataArray

# From the BARRA-C2 4.4 km downscaled reanalysis over Australian domain
da_list = {}
model_id = 'BARRA-C2'
domain_id = 'AUST-04'
da_list[model_id] = {}
# Hourly maximum, 2-metre temperature
da_list[model_id]['tasmax'] = open_barra_dataset(model_id, domain_id, freq, 'tasmax', yearmonth)['tasmax']
# Hourly 10-metre, zonal wind
da_list[model_id]['uas'] = open_barra_dataset(model_id, domain_id, freq, 'uas', yearmonth)['uas']
# Hourly 10-metre, meridional wind
da_list[model_id]['vas'] = open_barra_dataset(model_id, domain_id, freq, 'vas', yearmonth)['vas']

# From the BARRA-R2 12 km reanalysis over CORDEX-Australasia domain
model_id = 'BARRA-R2'
domain_id = 'AUS-11'
da_list[model_id] = {}
# Hourly mean-sea level pressure
da_list[model_id]['psl'] = open_barra_dataset(model_id, domain_id, freq, 'psl', yearmonth)['psl']
# Hourly 500 hPa zonal wind
da_list[model_id]['ua850'] = open_barra_dataset(model_id, domain_id, freq, 'ua850', yearmonth)['ua850']
# Hourly 500 hPa meridional wind
da_list[model_id]['va850'] = open_barra_dataset(model_id, domain_id, freq, 'va850', yearmonth)['va850']
````

````python
# Print their standard name and units
for model_id in da_list:
    for variable in da_list[model_id]:
        print(f"From {model_id}: {variable}: {da_list[model_id][variable].standard_name}, {da_list[model_id][variable].units}")
````

````python
# Visualise the BARRA-C2 screen-level air temperature and 10-metre wind.
# Screen-level (1.5-2-metre) temperature is a essential climate variable that directly 
#   informs extreme heat at human height. 
# 10-m wind can influence how extreme heat develops and persists. For instance, wind
#   winds allow heat build up, and wind direction determines whether a region receives
#   warm continental air intensifying heatwaves.

# Set time, in UTC, for plotting
valid_time = '2017-01-11 05:00'
# Define Southeastern Australian domain
se_aus_latrange = slice(-45, -20)
se_aus_lonrange = slice(135, 155)
model_id = 'BARRA-C2'
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(projection=ccrs.PlateCarree())

# Plot temperature
da_tas_sel = da_list[model_id]['tasmax'].sel(lat=se_aus_latrange, lon=se_aus_lonrange).sel(time=valid_time, method='nearest')
# Convert from Kelvin to degC
da_tas_sel = da_tas_sel - 273.15
da_tas_sel.plot(cmap='Reds', levels=np.linspace(20, 40, 10))

# Plot wind field
da_u_sel = da_list[model_id]['vas'].sel(lat=se_aus_latrange, lon=se_aus_lonrange).sel(time=valid_time, method='nearest')
da_v_sel = da_list[model_id]['uas'].sel(lat=se_aus_latrange, lon=se_aus_lonrange).sel(time=valid_time, method='nearest')
ds_uv = xr.merge([da_u_sel, da_v_sel])
nlat, nlon = da_v_sel.shape
# Thin the data horizontally to avoid having very densed wind vectors
ds_uv_subsampled = ds_uv.isel(lat=range(0, nlat, 20), lon=range(0, nlon, 20)).compute()
ds_uv_subsampled.plot.quiver(x='lon', y='lat', u='uas', v='vas', color='blue')

ax.coastlines()
ax.set_xlim(se_aus_lonrange.start, se_aus_lonrange.stop)
ax.set_ylim(se_aus_latrange.start, se_aus_latrange.stop)
ax.set_title(f"Near-surface Temperature (degC) and 10-metre Winds\nValid: {valid_time}");
````

This map shows local‑scale temperature variations and shows how they are shaped by surface winds and surrounding topography.

````python
# Define a larger Australian domain to look at broader synoptic environment
aus_latrange = slice(-47, -10)
aus_lonrange = slice(111, 180)
model_id = 'BARRA-R2'

fig = plt.figure(figsize=(15, 12))
ax = plt.subplot(projection=ccrs.PlateCarree())
from scipy.ndimage import gaussian_filter

# Plot MSLP
da_psl_sel = da_list[model_id]['psl'].sel(lat=aus_latrange, lon=aus_lonrange).sel(time=valid_time, method='nearest')
da_psl_sel = da_psl_sel / 100 # Convert Pa to hPa
psl_smooth = gaussian_filter(da_psl_sel, sigma=3)
levels = np.arange(psl_smooth.min(), psl_smooth.max(), 2)
cs = ax.contour(da_psl_sel['lon'], da_psl_sel['lat'], psl_smooth, levels=levels, colors='black', linewidths=1.2)
ax.clabel(cs, inline=True, fontsize=8)

# Plot wind field
da_u_sel = da_list[model_id]['va850'].sel(lat=aus_latrange, lon=aus_lonrange).sel(time=valid_time, method='nearest')
da_v_sel = da_list[model_id]['ua850'].sel(lat=aus_latrange, lon=aus_lonrange).sel(time=valid_time, method='nearest')
# Subsample to avoid clutter
step = 20
lon_sub = da_u_sel['lon'][::step]
lat_sub = da_u_sel['lat'][::step]
u_sub = da_u_sel.values[::step, ::step]
v_sub = da_v_sel.values[::step, ::step]
ax.barbs(lon_sub, lat_sub, 
         u_sub, v_sub,
         length=6, linewidth=0.8,
         color='darkgreen'
        )
# Add coastlines
ax.coastlines(color='blue')
# Set the plot domain
ax.set_xlim(aus_lonrange.start, aus_lonrange.stop)
ax.set_ylim(aus_latrange.start, aus_latrange.stop)
# Add title to figure
ax.set_title(f"MSLP (hPa) and 850 hPa Winds\nValid: {valid_time}");
````

This map shows the large-scale circulation patterns and air-mass characteristics that create and sustain extreme heat.

#### EXAMPLE 2: Brisbane Halloween Hailstorm October 2020

As part of a period of persistent thunderstorm activity across southeast Queensland from 24–31 October, a severe thunderstorm struck south-east Queensland, including Brisbane, on 31 October 2020. Hailstones up to 14 cm diameter were recorded, and the damage bill exceeded $1 billion. Nine supercell storms tracked across south-east Queensland during several hours. 

The representation of this event can be explored through various storm attribute fields, including precipitation and simulated radar reflectivity, updraft helicity, and several convective parameters.

````python
# Define which sub-collection of BARRA2
yearmonth = '202010'

# Open the BARRA2 datasets using xarray
#   and put them in a da_list dict as xr.DataArray

# From the BARRA-C2 4.4 km downscaled reanalysis over Australian domain
da_list = {}
model_id = 'BARRA-C2'
domain_id = 'AUST-04'
da_list[model_id] = {}
# Hourly precipitation
da_list[model_id]['pr'] = open_barra_dataset(model_id, domain_id, '1hr', 'pr', yearmonth)['pr']
# 20min simulated radar reflectivity at 1 km
da_list[model_id]['radrefl1km'] = open_barra_dataset(model_id, domain_id, '20min', 'radrefl1km', yearmonth)['radrefl1km']
# 20min updraft helicity
da_list[model_id]['helicitymax'] = open_barra_dataset(model_id, domain_id, '1hr', 'helicitymax', yearmonth)['helicitymax']

# From the BARRA-R2 12 km reanalysis over CORDEX-Australasia domain
model_id = 'BARRA-R2'
domain_id = 'AUST-11'
da_list[model_id] = {}
# Hourly most unstable CAPE
da_list[model_id]['MUCAPE'] = open_barra_dataset(model_id, domain_id, freq, 'MUCAPE', yearmonth)['MUCAPE']
````

````python
# Print their standard name and units
for model_id in da_list:
    for variable in da_list[model_id]:
        print(f"From {model_id}: {variable}: {da_list[model_id][variable].standard_name}, {da_list[model_id][variable].units}")
````

````python
# Plot precipitation at different times during the event
valid_times = ['2020-10-31 02:30', '2020-10-31 04:30', '2020-10-31 06:30', '2020-10-31 08:30', '2020-10-31 10:30', '2020-10-31 12:30']

# Define Brisbane domain
latrange = slice(-30, -23)
lonrange = slice(149, 155)
model_id = 'BARRA-C2'

# variable name to plot
variable = 'pr'
# figure label
label = 'Hourly precipitation (mm/h)'

# Define discrete precipitation bins (mm/hr)
bounds = [0, 0.2, 1, 3, 10, 20, 40]
# Custom discrete colors (light → heavy)
colors = [
    "#f2f2f2",  # 0–0.2 drizzle
    "#9ecae1",  # 0.2–1 light blue
    "#3182bd",  # 1–3 blue
    "#31a354",  # 3–10 green
    "#fed976",  # 10–40 yellow
    "#b10026"   # 40–80 dark red (extreme)
]
cmap = ListedColormap(colors)
norm = BoundaryNorm(bounds, ncolors=cmap.N)

fig, axes = plt.subplots(2, 3, figsize=(12, 8), subplot_kw={"projection": ccrs.PlateCarree()})

for ax, valid_time in zip(axes.flat, valid_times):
    # Subset the data for domain and time
    da = da_list[model_id][variable].sel(lat=latrange, lon=lonrange).sel(time=valid_time)
    # convert from mm/s to mm/h
    da = da * (60*60)
    # Plot
    da.plot(
        ax=ax,
        cmap=cmap, 
        norm=norm,
        extend='max',
        cbar_kwargs={'shrink': 0.5, 'label':''}
    )

    # Add coastlines
    ax.coastlines()
    # Set the plot domain
    ax.set_xlim(lonrange.start, lonrange.stop)
    ax.set_ylim(latrange.start, latrange.stop)
    # Add title to each subfigure
    ax.set_title(f"{label}\nValid: {valid_time}")
````

Extreme precipitation is often highly localized, and rain‑gauge networks are uneven. 
Reanalysis fills these gaps by blending observations with a weather model to produce continuous, 
gridded precipitation estimates. 
Here BARRA-C2 shows the possible locations and times of the heavy precipitation and hail.

Updraft helicity is a useful diagnostic to identify supercell-like structures in convection-permitting model simulations. 
This diagnostic is also unique to convection-permitting models such as BARRA-C2. 
It is available as hourly minimum and maximum fields, integrated from 2–5 km AGL, which capture 
cyclonic and anti-cyclonic rotating storms, respectively. The absolute maximum of these two fields can 
be used to identify storms rotating in either sense. When aggregated over multiple hours, the resulting 
UH "swaths" provide an indication of the location, intensity, and evolution of modelled severe storms. 
For simplicity, here we consider the hourly maximum fields.

````python
# Plot most-unstable CAPE at different times
valid_times = ['2020-10-31 02:00', '2020-10-31 04:00', '2020-10-31 06:00', '2020-10-31 08:00', '2020-10-31 10:00', '2020-10-31 12:00']
model_id = 'BARRA-R2'
variable = 'MUCAPE'
label='Hourly most unstable CAPE (J kg-1)'

fig, axes = plt.subplots(2, 3, figsize=(12, 8), subplot_kw={"projection": ccrs.PlateCarree()})

for ax, valid_time in zip(axes.flat, valid_times):
    # Subset the data over domain and time
    da = da_list[model_id][variable].sel(lat=latrange, lon=lonrange, time=valid_time)
    # Plot
    da.plot(
        ax=ax,
        vmin=0, vmax=2500,
        cmap='jet',
        cbar_kwargs={'shrink': 0.5, 'label':''}
    )
    
    # Add coastlines
    ax.coastlines(color='w')
    # Set the plot domain
    ax.set_xlim(lonrange.start, lonrange.stop)
    ax.set_ylim(latrange.start, latrange.stop)
    # Add title to each subfigure
    ax.set_title(f"{label}\nValid: {valid_time}")
````

BARRA-R2 does not simulate convective process explicitly to allow detailed study of convective storms or supercells.
However it provides diagnosis of the convective environment using measures such as most unstable Convective Available Potential Energy (CAPE).
CAPE is an indicator of atmospheric instability and is valuable for predicting categories of severe weather.

### Q: What is one question that you have answered using these data? Can you show us how you came to that answer?

A regional reanalysis is an exceptionally versatile weather and climate dataset that can help answering a wide spectrum of scientific and applied questions. At its core, it gives a physically consistent reconstruction of past atmospheric conditions, enabling detailed examination of climate extremes, everyday variability and long-term changes across the entire reanalysis domain. This level of spatial and temporal completeness goes beyond what can be achieved with observations alone, especially in regions with sparse measurement networks. Reanalyses also serve as input data for downstream modelling systems and as training data for AI/ML applications.

Examples of applications are already emerging in the published literature:
* [Pepler & Dowdy (2025)](https://doi.org/10.1175/JCLI-D-25-0022.1) analysed the frequency of co-occurring extreme rain and extreme wind (compound events) across Australia using BARRA2.
* [Dong et al. (2026)](https://doi.org/10.1016/j.coastaleng.2025.104892) used BARRA2 to drive a 45-year high-resolution wave hindcast for the Australian coastline, improving understanding of historical wave climate.
* [McGloin et al. (2026)](https://doi.org/10.1038/s44304-026-00193-9) employed BARRA2 as calibration data for bias correcting climate projections in their assessment of future changes in extreme fire weather conditions in Australia.
* [Palmer et al. (2025)](https://doi.org/10.1071/ES24028) evaluated the suitability of BARRA2 for wind-energy modelling to support techno-economic assessments of onshore wind-power potential.

### Q: What is one unanswered question that you think could be answered using these data? Do you have any recommendations or advice for someone wanting to answer this question?

One potential question is, *how do compound climate extremes—such as heat, drought, heavy rainfall, damaging winds, fire weather, storm surge, coastal inundation, and riverine flooding—co‑evolve at local scales, and how have their frequencies, spatial patterns, and underlying drivers changed over recent decades?*

Compound extremes are increasingly recognised as high‑impact hazards, yet they remain poorly quantified in many regions because observational networks are sparse and often lack the spatial completeness needed to capture co‑occurring events. A high‑resolution regional reanalysis such as BARRA2 provides the consistent, multivariate, multi‑decadal dataset required to examine these interactions in detail. It also supplies boundary and forcing data for hydrological, coastal, and wave models, enabling integrated assessments of hydrological and coastal hazards alongside atmospheric extremes.
